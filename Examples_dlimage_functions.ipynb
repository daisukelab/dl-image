```python
from dlcliche.notebook import *
from dlcliche.utils import *
from dlimage.functions import abstract_image

import urllib.request


# Load test images
org_files = []
for url, fname in [
    ['https://github.com/tensorflow/models/raw/master/research/object_detection/test_images/image1.jpg', '/tmp/image1.jpg'],
    ['https://github.com/tensorflow/models/raw/master/research/object_detection/test_images/image2.jpg', '/tmp/image2.jpg'],
]:
    urllib.request.urlretrieve(url, fname)
    org_files.append(fname)
```

```python
def test_abstract_image(files=org_files):
    test_params = [
        {'K': 10, 'grayscale': True, 'add_edge': True, 'resize': (256, 256)},
        {'K': 5, 'grayscale': False, 'add_edge': False, 'resize': (378, 256)},
        {'K': None, 'grayscale': False, 'add_edge': False, 'resize': (256, 256)},
        {'K': 5, 'grayscale': True, 'add_edge': True, 'resize': (512, 512)},
        {'K': 100, 'grayscale': False, 'add_edge': True, 'resize': (512, 512)},
    ]
    for f in files:
        for params in test_params:
            img, abst = abstract_image(f, **params)
            img = np.concatenate((img, abst), axis=1)
            fig = plt.figure(figsize=(12, 6))
            plt.imshow(img)
            plt.title(str(params))
            plt.axis('off')
            plt.show()


test_abstract_image()
```
